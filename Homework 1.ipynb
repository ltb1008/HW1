```python

#Problem 6
import numpy as np
import matplotlib.pyplot as plt
def func(x):
    return 0.1*x**3 - x**2 + 5
x = np.linspace(-5, 5, 400)
y = func(x)
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('func(x)')
plt.title('Plot of func(x)')
plt.annotate('Maximum', xy=(x[np.argmax(y)], max(y)), xytext=(3, 10),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Minimum', xy=(x[np.argmin(y)], min(y)), xytext=(-4, 0),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.grid(True)
plt.show()


print(x[np.argmax(y)], max(func(x)))
print(x[np.argmin(y)], min(func(x)))
```
